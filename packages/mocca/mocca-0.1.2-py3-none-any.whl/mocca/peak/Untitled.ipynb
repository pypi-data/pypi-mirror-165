```python
plt.clf()
plt.style.use('default')
ax = carbonate_df.plot(x='time_hrs', y='substrate_area_normed',marker = '.', label = '1-cyanocyclohexyl) methyl carbonate', markersize=12)
cn1oacethane_df.plot(x='time_hrs', y='substrate_area_normed', marker = '.', markersize=12, label = '1-cyanoethyl acetate', ax=ax)
bzocnethane_df.plot(x='rxn_times', y='normed_integral', marker = '.', markersize=12, label = '1-cyanoethyl benzoate', ax=ax)
cn2oacpropane_df.plot(x='time_hrs', y='substrate_area_normed', marker = '.', markersize=12, label = '(1-cyano-1-methyl-ethyl) acetate', ax=ax)
bzo2cnpropane_df.plot(x='rxn_times', y='normed_integral', marker = '.', markersize=12, label = '(1-cyano-1-methyl-ethyl) benzoate', ax=ax)

plt.xlabel('time (h)')
plt.ylabel('normed intensity')

plt.show()
```
